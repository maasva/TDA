```python
import os
import numpy as np
import datetime
```

```python
def createtime(date,time):
    year, month, day = date.split('-')
    hour, minute, second = time.split(':')
    return datetime.datetime(int(year), int(month), int(day), int(hour), int(minute), int(second))
```

```python
main_dir = f'Geolife Trajectories 1.3/Data/' 
numbers = os.listdir(main_dir)
file_avgs = []
for folder_number in numbers:
    directory = f'Geolife Trajectories 1.3/Data/{folder_number}/Trajectory/' 
    alltrajectories = os.listdir(directory)
    filelist = []
    for file in alltrajectories:
        trajectory = np.genfromtxt(directory+file,
                       skip_header=6, delimiter=',', dtype=None, encoding=None)
        skip = False
        timelist = []
        for row in trajectory:
            if skip == False: # skip the first row
                skip = True
                curTime = createtime(row[5], row[6])
                continue
            prevTime = curTime
            curTime = createtime(row[5], row[6])
            timeDiff = curTime-prevTime
            timelist.append(timeDiff.total_seconds())
        filelist.append(sum(timelist)/len(timelist))
    file_avgs.append(sum(filelist)/len(filelist))
    print('did files', folder_number) 
overall_avg = sum(file_avgs)/len(file_avgs)
```

```python
main_dir = f'Filtered_data/' 
people = os.listdir(main_dir)
maxdate = datetime.datetime(1,1,1)
mindate = datetime.datetime(3000,1,1)
list_of_trajectories = []
for person in people:
    trajectory = np.genfromtxt(main_dir+person,
                       skip_header=0, delimiter=',', dtype=None, encoding=None)
    if trajectory.ndim == 0 or trajectory.size == 0:
        continue
    list_of_trajectories.append(trajectory)
    first = trajectory[0]
    last = trajectory[-1]
    earlydate = createtime(first[2], first[3])
    latedate = createtime(last[2], last[3])
    if earlydate <mindate:
        mindate = earlydate
    if latedate > maxdate:
        maxdate = latedate
print(maxdate)
print(mindate)
```

```python
date = datetime.datetime(2007,4,12,9,31,0) # we round down the mindate to minutes
accumulated_data = {}
while date <= maxdate:
    Data = [None for i in range(len(list_of_trajectories))]
    for i, trajectory in enumerate(list_of_trajectories):
        for j, row in enumerate(trajectory):
            cur_date = createtime(row[2], row[3])
            if cur_date <= date: 
                Data[i] = (row[0], row[1]) # set the location
                list_of_trajectories[i] = np.delete(trajectory, range(j+1),0)
            else: # if the statement is not true for this row, it won't be for later rows, so we can break the loop
                break
    date += datetime.timedelta(minutes=1)
    if Data !=  [None for i in range(len(list_of_trajectories))]:
        accumulated_data[date] = Data
```

```python
import pandas as pd
df = pd.DataFrame(accumulated_data)
```
